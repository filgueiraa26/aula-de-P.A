```python
# Calculadora Simples

# Entrada 2 valores
# Processamento calcuras usando operadores aritiméticos
#  Saída Mostraro resultado


# Entrada

numero1 = int (input("digite um número: "))
numero2 = int (input("digite o segundo número: "))


# Processamento

soma = numero1 +numero2
subtração = numero1 - numero2
multiplicação = numero1 * numero2
divisão = numero1 / numero2
div_inteira = numero1 // numero2
resto = numero1 % numero2
potência = numero1 ** numero2

# Saída


print (f"o resultado da soma de dois números é igual a:  {soma}")
print (f"o resultado da subtração de dois númeors é igual a:  {subtração}")
print (f"o resultado da multiplicação de dois números é igual a: {multiplicação}")
print (f" o resultado da divisão de dois números é igual a: {divisão}")
print (f" o resultado da divisão inteira de dois números é igual a: {div_inteira}")
print (f" o resto da divisão de dois números é igual a: {resto}")
print (f" o resultado da potência de dois números é igual a: {potência}")
```
